```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
path=r'C:\Users\unu\Downloads\insurance.csv'
df=pd.read_csv(path)
#check for nulls in df
df.isnull().sum
#encode the data that is not numbers
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['smoker']=le.fit_transform(df['smoker'])
df['region']=le.fit_transform(df['region'])
x=df.drop(columns='charges',axis=1)
y=df['charges']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_train)
r2_train=metrics.r2_score(y_train,y_pred)
x_pred=reg.predict(x_test)
r2_test=metrics.r2_score(y_test,x_pred)
print(r2_train)
print(r2_test)
```
